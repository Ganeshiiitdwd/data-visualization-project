```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('crime_data.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
# Categorize crimes
violent_crimes = df[['Murder', 'Rape', 'Riots']].sum().sum()
non_violent_crimes = df[['Theft', 'Cheating', 'Forgery']].sum().sum()

# Data for pie chart
crime_data = [violent_crimes, non_violent_crimes]
labels = ['Violent Crimes', 'Non-Violent Crimes']

# Create pie chart
plt.figure(figsize=(7, 7))
plt.pie(crime_data, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'], explode=(0.1, 0))

# Add title
plt.title('Distribution of Violent vs Non-Violent Crimes ', fontsize=16)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Sum the total crimes for each State/UT by grouping
df_statewise = df.groupby('States/UTs')['Total Cognizable IPC crimes'].sum().reset_index()

# Sort the data based on total crimes in descending order
df_statewise_sorted = df_statewise.sort_values('Total Cognizable IPC crimes', ascending=False)

# Set the style for aesthetic appeal
sns.set(style="whitegrid")

# Create a barplot with states on the y-axis, total crimes on the x-axis
plt.figure(figsize=(10, 8))
sns.barplot(x='Total Cognizable IPC crimes', y='States/UTs', data=df_statewise_sorted, palette='Blues_r')

# Add titles and labels
plt.title('Total Cognizable IPC Crimes by States/UTs ', fontsize=16)
plt.xlabel('Total Crimes', fontsize=12)
plt.ylabel('States/UTs', fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
